# 넘파이 사용하기

```python
import numpy as np

a=np.array([1,2,3,4,5])
b=np.array(['대한민국', '포르투갈', '가나', '우루과이'])
c=np.array([1,2,'대한민국', '포르투갈'])

print(a)
print(b)
print(c)
```

# 판다스 사용하기

```python
import pandas as pd

s=pd.Series(['대한민국', '포르투갈', '가나', '우루과이'], index=['가', '나', '다', '라'], name="2022 카타르월드컵 H조")

print(s)
```
